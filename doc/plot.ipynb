```python
import pandas as pd
import numpy as np
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import matplotlib.pyplot as plt
```

```python
range_data = 100
x = np.arange(0, range_data, 1)/10
mapx = np.interp(x, (0, range_data), (0, 10))
y_title = 1.05

with open("kp4.txt", "r") as file:
    next(file)
    
    # Read the first 100 lines
    data4 = [next(file) for _ in range(range_data)]
with open("kpp5.txt", "r") as file:
    next(file)
    
    # Read the first 100 lines
    data5 = [next(file) for _ in range(range_data)]
with open("kp6.txt", "r") as file:
    next(file)
    
    # Read the first 100 lines
    data6 = [next(file) for _ in range(range_data)]
with open("kp7.txt", "r") as file:
    next(file)
    
    # Read the first 100 lines
    data7 = [next(file) for _ in range(range_data)]
with open("kp8.txt", "r") as file:
    next(file)
    
    # Read the first 100 lines
    data8 = [next(file) for _ in range(range_data)]
with open("kp9.txt", "r") as file:
    next(file)
    
    # Read the first 100 lines
    data9 = [next(file) for _ in range(range_data)]
with open("kp10.txt", "r") as file:
    next(file)
    
    # Read the first 100 lines
    data10 = [next(file) for _ in range(range_data)]
with open("kp20.txt", "r") as file:
    next(file)
    
    # Read the first 100 lines
    data20 = [next(file) for _ in range(range_data)]
with open("kp50.txt", "r") as file:
    next(file)
    
    # Read the first 100 lines
    data50 = [next(file) for _ in range(range_data)]


# Extracting setpoint and yaw values from the data
setpoints    = [float(line.split(',')[0]) for line in data4]
yaw_values4  = [float(line.split(',')[1]) for line in data4]
yaw_values5  = [float(line.split(',')[1]) for line in data5]
yaw_values6  = [float(line.split(',')[1]) for line in data6]
yaw_values7  = [float(line.split(',')[1]) for line in data7]
yaw_values8  = [float(line.split(',')[1]) for line in data8]
yaw_values9  = [float(line.split(',')[1]) for line in data9]
yaw_values10 = [float(line.split(',')[1]) for line in data10]
yaw_values20 = [float(line.split(',')[1]) for line in data20]
yaw_values50 = [float(line.split(',')[1]) for line in data50]

fig, axs = plt.subplots(6, figsize=(18,15), sharex=True, sharey=True)
fig.suptitle('Grafik Respon Mobile Robot (Tuning PID)', y=0.91)

axs[0].plot(x, setpoints, linestyle='-', color='r')
axs[0].plot(x, yaw_values4, linestyle='-', color='m')
axs[0].set_title('Kp = 4',  y=y_title, pad=-20)
axs[0].set_ylabel("Sudut ()")

axs[1].plot(x, setpoints, linestyle='-', color='r')
axs[1].plot(x, yaw_values5, linestyle='-', color='b')
axs[1].set_title('Kp = 5',  y=y_title, pad=-20)
axs[1].set_ylabel("Sudut ()")

axs[2].plot(x, setpoints, linestyle='-', color='r')
axs[2].plot(x, yaw_values6, linestyle='-', color='y')
axs[2].set_title('Kp = 6',  y=y_title, pad=-20)
axs[2].set_ylabel("Sudut ()")

axs[3].plot(x, setpoints, linestyle='-', color='r')
axs[3].plot(x, yaw_values8, linestyle='-', color='g')
axs[3].set_title('Kp = 8',  y=y_title, pad=-20)
axs[3].set_ylabel("Sudut ()")

axs[4].plot(x, setpoints, linestyle='-', color='r')
axs[4].plot(x, yaw_values9, linestyle='-', color='c')
axs[4].set_title('Kp = 9',  y=y_title, pad=-20)
axs[4].set_ylabel("Sudut ()")

axs[5].plot(x, setpoints, linestyle='-', color='r')
axs[5].plot(x, yaw_values10, linestyle='-', color='olive')
axs[5].set_title('Kp = 10',  y=y_title, pad=-20)
axs[5].set_xlabel("Waktu (s)")
axs[5].set_ylabel("Sudut ()")

# Adding legends
axs[0].legend(['setpoint', 'yaw'])
axs[1].legend(['setpoint', 'yaw'])
axs[2].legend(['setpoint', 'yaw'])
axs[3].legend(['setpoint', 'yaw'])
axs[4].legend(['setpoint', 'yaw'])
axs[5].legend(['setpoint', 'yaw'])

axs[3].xaxis.set_major_locator(MultipleLocator(1))
axs[0].yaxis.set_major_locator(MultipleLocator(5))
axs[0].yaxis.set_minor_locator(MultipleLocator(5))
axs[1].yaxis.set_major_locator(MultipleLocator(5))
axs[1].yaxis.set_minor_locator(MultipleLocator(5))
axs[2].yaxis.set_major_locator(MultipleLocator(5))
axs[2].yaxis.set_minor_locator(MultipleLocator(5))
axs[3].yaxis.set_major_locator(MultipleLocator(5))
axs[3].yaxis.set_minor_locator(MultipleLocator(5))
axs[4].yaxis.set_major_locator(MultipleLocator(5))
axs[4].yaxis.set_minor_locator(MultipleLocator(5))
axs[5].yaxis.set_major_locator(MultipleLocator(5))
axs[5].yaxis.set_minor_locator(MultipleLocator(5))

# plt.xticks(x)
axs[0].grid()
axs[1].grid()
axs[2].grid()
axs[3].grid()
axs[4].grid()
axs[5].grid()
plt.subplots_adjust(hspace=0)
plt.show()
```
